# Regular expression

## 常见符号含义

在Python中使用`re.search`函数进行正则表达式搜索时，常见的符号及其含义如下：

1. `.` - 匹配任意单个字符，除了换行符`\n`。
2. `^` - 匹配字符串的开头。
3. `$` - 匹配字符串的结尾。
4. `*` - 匹配前面的子表达式零次或多次。
5. `+` - 匹配前面的子表达式一次或多次。
6. `?` - 匹配前面的子表达式零次或一次，或指示非贪婪匹配（尽量匹配最少的字符）。
7. `{}` - 用来指定匹配的次数，例如`{3}`表示匹配3次，`{3,5}`表示匹配3到5次。
8. `[]` - 用来指定字符集合，可以匹配其中任意一个字符。
9. `()` - 用来创建一个子组，可以对其进行后向引用。
10. `|` - 用来表示或操作，匹配符号左右任意一个表达式。
11. `\` - 转义字符，用来取消紧跟其后的字符的特殊含义。
----
1. **`\w`**: 匹配任何单字字符（字母、数字、下划线），等同于 `[a-zA-Z0-9_]`。
2. **`\W`**: 匹配任何非单字字符，等同于 `[^a-zA-Z0-9_]`。
3. **`\d`**: 匹配任何数字字符，等同于 `[0-9]`。
4. **`\D`**: 匹配任何非数字字符，等同于 `[^0-9]`。
5. **`\s`**: 匹配任何空白字符，包括空格、制表符、换行符等，等同于 `[ \t\r\n\f]`。
6. **`\S`**: 匹配任何非空白字符，等同于 `[^ \t\r\n\f]`。
7. **`.`**: 匹配除换行符之外的任何单个字符。
8. **`^`**: 匹配输入的开始位置。
9. **`$`**: 匹配输入的结束位置。
10. **`[]`**: 匹配方括号内的任何字符。例如，`[abc]` 匹配 `a`、`b` 或 `c`。
11. **`[^]`**: 匹配不在方括号内的任何字符。例如，`[^abc]` 匹配任何不是 `a`、`b` 或 `c` 的字符。
12. **`|`**: 匹配左边或右边的表达式。例如，`a|b` 匹配 `a` 或 `b`。
13. **`*`**: 匹配前一个元素零次或多次。
14. **`+`**: 匹配前一个元素一次或多次。
15. **`?`**: 匹配前一个元素零次或一次，或者用于非贪婪匹配。
16. **`{n}`**: 匹配前一个元素恰好 `n` 次。
17. **`{n,}`**: 匹配前一个元素至少 `n` 次。
18. **`{n,m}`**: 匹配前一个元素至少 `n` 次，但不超过 `m` 次。
19. **`()`**: 分组，将多个字符作为一个单元进行匹配。

## Python中的使用

### re.sub(pattern, repl, string, count=0, flags=0) 替换所有匹配项为制定内容

```python
text = "<div><td valign=\"top\">Moo</td></div>"
pattern = r"<[^>]+>"
re.sub(pattern,"",text)
```

### re.findall() 找出所有匹配的元素,以列表形式返回

```python
text = "My social security number is 123-45-6789 bro, or actually maybe it's 321-45-6789."
pattern = r"[0-9]{3}-[0-9]{2}-[0-9]{4}"
re.findall(pattern,text)
```

### 表达式中用"()"圈起来的会被返回

```python
text = "observations : 03:04:53 - Horse awakens. 03:05:14 - Horse goes back to sleep."
pattern = r"(\d\d):(\d\d):(\d\d) - ([\w+ \w]+)"
matches = re.findall(pattern,text)
matches
```

### re.match(pattern,string) 匹配
从字符串的起始位置开始匹配正则表达式。如果正则表达式匹配成功，返回一个匹配对象；否则返回 None。

```python
result = re.match(r'\d+', '123abc')
print(result.group())  # 输出: 123
```

### re.search(pattern, string) 匹配第一次出现的
扫描整个字符串，查找第一个匹配正则表达式的子串。如果匹配成功，返回一个匹配对象；否则返回 None。

```python
result = re.search(r'\d+', '123abc123')  
print(result.group())  # 输出: 123
```

### re.split(pattern string) 返回一个符合匹配的列表
根据正则表达式分割字符串，返回一个列表。虽然它不是直接提取匹配内容，但它可以帮助你提取被分隔符分开的部分。

```python
result = re.split(r'\W+', 'Hello, World! Python is great.')
print(result)  # 输出: ['Hello', 'World', 'Python', 'is', 'great', '']
```

## 海象运算符
:= 的作用是执行赋值操作，并返回赋值的结果

```python
name = input("what's your name?").strip()
if matches := re.search(r"^(.+), *(.+)$",name):
  name = matches.group(2)+" "+matches.group(1)
print(f"hello,{name}")
```
