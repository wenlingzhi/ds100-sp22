```python
text = "I have 10 apples, GREAT!"
pattern = r"[a-z0-9]+"
result = re.findall(pattern,text)
result
```

```python
text = "I've got 10 eggs, 20 gooses, and 30 giants."
pattern = r"[0-9]{2} [a-z]+"
result = re.findall(pattern,text)
result
```

```python
text = "abc, 123"
pattern = r"[a-z1,9]+"
result = re.findall(pattern,text)
result
```
